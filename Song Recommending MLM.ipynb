```python
from sklearn.model_selection import train_test_split
train_data , test_data = train_test_split(music_features_scaled, test_size = 0.3 , random_state = 42)
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
inertia = []
k_values = range(1,6)

for k in k_values:
  kmeans = KMeans(n_clusters=k, random_state=42)
  kmeans.fit(train_data)
  inertia.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(k_values, inertia, marker='o', linestyle='-', color='b')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

for k in range(2, 5):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(train_data)
    print(f"K={k}, Silhouette Score={silhouette_score(train_data, labels):.3f}")
```

```python
optimal_k = 4
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df['cluster'] = kmeans.fit_predict(music_features_scaled)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend_songs(song_name, df=df, num_recommendation=5):
    # Strip whitespace from the input song name
    song_name = song_name.strip()

    # Filter the DataFrame for the given song name
    song_info = df[df['Title'] == song_name]

    # Check if the song was found
    if song_info.empty:
        print(f"Song '{song_name}' not found in the dataset.")
        return pd.DataFrame() # Return an empty DataFrame if song not found

    # Get the cluster of the song
    song_cluster = song_info['cluster'].values[0]

    # Filter for songs in the same cluster and sample
    # Ensure we don't recommend the same song
    similar_songs = df[df['cluster'] == song_cluster].sample(num_recommendation + 1, replace=False)
    similar_songs = similar_songs[similar_songs['Title'] != song_name]

    # If after removing the original song, we have fewer than num_recommendation, sample again
    if len(similar_songs) < num_recommendation:
         similar_songs = df[df['cluster'] == song_cluster].sample(num_recommendation, replace=False)
         similar_songs = similar_songs[similar_songs['Title'] != song_name]


    return similar_songs[['Title', 'Artist', 'cluster']] # Return relevant columns
```

```python
def recommend_songs_2(song_name, df=df, num_recommendation=5):
  # Corrected column name from 'name' to 'Title'
  song_cluster = df[df['Title']==song_name]['cluster'].values[0]
  same_cluster_song = df[df['cluster']==song_cluster]

  # Corrected typo in function name cosine_similatriy to cosine_similarity
  song_index = same_cluster_song[same_cluster_song['Title'] == song_name].index[0] # Use 'Title' here too
  cluster_features = same_cluster_song[audio_features]

  similarity = cosine_similarity(cluster_features,cluster_features)

  similar_songs = np.argsort(similarity[song_index])[::-1][1:num_recommendation+1]

  return same_cluster_song.iloc[similar_songs]
```
